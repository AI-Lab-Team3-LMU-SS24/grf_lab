### Simple example of fitting an NVP flow to a 2D dataset

Paper is [here](https://arxiv.org/pdf/1605.08803).

```python
from copy import deepcopy
import os
import torch
from torch import nn
from torch import distributions
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from tqdm.notebook import trange
from typing import Optional, Union

FIGS_DIR = os.path.join(os.getcwd(), "figs/moons/")
if not os.path.exists(FIGS_DIR):
    os.makedirs(FIGS_DIR, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
def convert(x):
    # Convert a numpy array to a torch tensor with float32 entries
    return torch.as_tensor(x).float()

device
```

##### Model

```python
class RealNVP(nn.Module):
    def __init__(self, net_s, net_t, mask, prior):
        super(RealNVP, self).__init__()
        # Base distribution, a data-dimensional Gaussian
        self.prior = prior
        # Masks are not to be optimised
        self.mask = nn.Parameter(mask, requires_grad=False)
        # The s and t nets that parameterise the scale and shift
        # change of variables according to inputs, here we are
        # duplicating the networks for each layer.
        self.t = torch.nn.ModuleList(
            [net_t() for _ in range(len(masks))]
        )
        self.s = torch.nn.ModuleList(
            [net_s() for _ in range(len(masks))]
        )

    def reverse(self, z):
        # Map from Gaussian distributed z to data x
        x = z
        for i in range(len(self.t)):
            x_ = x * self.mask[i]
            s = self.s[i](x_) * (1 - self.mask[i])
            t = self.t[i](x_) * (1 - self.mask[i])
            x = x_ + (1 - self.mask[i]) * (x * torch.exp(s) + t)
        return x

    def forward(self, x):
        # Map from data x to Gaussian distributed z
        log_det_J, z = x.new_zeros(x.shape[0]), x
        for i in reversed(range(len(self.t))):
            z_ = self.mask[i] * z
            s = self.s[i](z_) * (1 - self.mask[i])
            t = self.t[i](z_) * (1 - self.mask[i])
            z = (1 - self.mask[i]) * (z - t) * torch.exp(-s) + z_
            log_det_J -= s.sum(dim=1)
        return z, log_det_J

    def log_prob(self, x):
        # Calculate log-probability of x under flow distribution
        z, log_det_J = self.forward(x)
        return self.prior.log_prob(z) + log_det_J

    def sample(self, n):
        # Sample n points from flow distribution fit to data
        z = self.prior.sample((n, 1))
        x = self.reverse(z)
        return x
```

**Incorparating conditional data**

We want to find $\mathbb{P}(x | c)$ using the above flow-model.

$\mathbb{P}(x | c) = \frac{\mathbb{P}(c | x) \cdot \mathbb{P}(x)}{\mathbb{P}(c)} = \frac{\mathbb{P}(c \,\cap\, x)}{\mathbb{P}(c)}$

$\mathbb{P}(c)$ can be easily calculated by calculating the class ratio
and if we train the fllow model over both $x$ and $c$, it should learn their joint probability $\mathbb{P}(c \,\cap\, x)$, which is enough to calculate $\mathbb{P}(x | c)$.

##### Setup

```python
H = 256            # Hidden dimension
L = 4              # Number of transforms in flow
n_data = 10000     # Number of datapoints in training set
n_steps = 10000    # Number of batches to run for
batch_size = 1000  # Number of datapoints in batch
inp_params = 3     # x, y, c

# Scaling and shifting networks for flow change of variables
net_s = lambda: nn.Sequential(
    nn.Linear(inp_params, H), nn.LeakyReLU(),
    nn.Linear(H, H), nn.LeakyReLU(),
    nn.Linear(H, inp_params), nn.Tanh()
)
net_t = lambda: nn.Sequential(
    nn.Linear(inp_params, H), nn.LeakyReLU(),
    nn.Linear(H, H), nn.LeakyReLU(),
    nn.Linear(H, inp_params)
)

# Masking for NVP flow model
masks = torch.from_numpy(np.array([[1, 0, 1], [0, 1, 0]] * L)).float()

# Base distribution for normalizing flow
prior = distributions.MultivariateNormal(
    torch.zeros(inp_params).to(device), torch.eye(inp_params).to(device)
)

# Build the model
flow = RealNVP(net_s, net_t, masks, prior).to(device)

# Get some training data
X, Y = make_moons(n_data, noise=0.05)

# Scale the data to [-1, 1] 
X = 2. * (X - X.min()) / (X.max() - X.min()) - 1.
Y = 2 * Y - 1
data = np.hstack([X, Y.reshape(-1, 1)])

# Convert it to tensors and move to device
X, Y, data = convert(X).to(device), convert(Y).to(device), convert(data).to(device)

def get_data(n_batch):
    # A simple dataloader
    ix = torch.randint(low=0, high=n_data, size=(n_batch,))
    return X[ix], Y[ix], data[ix]

optimizer = torch.optim.Adam(flow.parameters(), lr=1e-4)
```

```python
class DataGenerator:
    def __init__(self, n_data: int, validation_size: Optional[Union[float, int]] = None):
        self.size = n_data
        if isinstance(validation_size, float):
            assert validation_size > 0 and validation_size < 1
            validation_size = int(validation_size * n_data)
            assert validation_size > 0 and validation_size < n_data
        self.validation_size = validation_size
        self.validation_frac = validation_size / n_data
        X, Y = make_moons(n_data, noise=0.05)
    
        # Scale the data to [-1, 1] 
        X = 2. * (X - X.min()) / (X.max() - X.min()) - 1.
        Y = 2 * Y - 1
        data = np.hstack([X, Y.reshape(-1, 1)])
        
        # Convert it to tensors and move to device
        X, Y, data = convert(X).to(device), convert(Y).to(device), convert(data).to(device)
        if self.validation_size is None:
            self.x_train, self.y_train, self.data_train = X, Y, data
            self.x_vald, self.y_vald, self.data_vald = None, None, None
            return
        self.x_train, self.y_train, self.data_train = X[self.validation_size:], Y[self.validation_size:], data[self.validation_size:]
        self.x_vald, self.y_vald, self.data_vald = X[:self.validation_size], Y[:self.validation_size], data[:self.validation_size]

    def __call__(self, batch_size: int):
        # A simple dataloader
        idx_train = torch.randint(low=0, high=self.size, size=(batch_size,))
        if self.validation_size is None:
            return X[idx_train], Y[idx_train], data[idx_train]
        idx_vald = torch.randint(low=0, high=self.validation_size, size=(int(batch_size * self.validation_frac),))
        return (X[idx_train], Y[idx_train], data[idx_train]), (X[idx_vald], Y[idx_vald], data[idx_vald])
```

```python
# Plot some training samples for comparison
fig, ax = plt.subplots(1, 1, dpi=200)
ax.scatter(*X.cpu().numpy().T, c=Y.cpu().numpy(), s=1., cmap="bwr")
plt.tight_layout()
plt.savefig(os.path.join(FIGS_DIR, "train_samples_moons.png"))
plt.show()
```

#### Train

```python
losses_train, losses_vald = list(), list()
dg = DataGenerator(n_data, 0.2)
early_stopping = 240

stop_counter = 0
best_loss_vlad = 1_000_000.
best_state = flow.state_dict()
with trange(n_steps) as T:
    # model.set_classes_prob(Y)
    for t in T:
        train, vald = dg(batch_size)
        # _, _, x = get_data(batch_size)
        _, _, x_train = train
        _, _, x_vald = vald
        x_train = x_train.to(device)
        x_vald = x_vald.to(device)

        flow.train(True)
        loss_train = -flow.log_prob(x_train).mean()
        optimizer.zero_grad()
        loss_train.backward()
        optimizer.step()
        losses_train.append(loss_train.item())

        flow.eval()
        with torch.no_grad():
            loss_vald = -flow.log_prob(x_vald).mean()
            losses_vald.append(loss_vald.item())
        
        T.set_postfix(loss=f"{loss_train:.3E}")

        stop_counter += 1
        if loss_vald < best_loss_vlad:
            best_loss_vlad = loss_vald
            best_state = deepcopy(flow.state_dict())
            stop_counter = 0

        if stop_counter >= early_stopping:
            break

        if t % 500 == 0:
            # Sample from the model
            samples = flow.sample(2000)
            samples = samples.cpu().detach().numpy()

            fig, ax = plt.subplots(1, 1, dpi=100)
            ax.scatter(*samples[..., :2].T, s=1.)
            plt.tight_layout()
            # plt.savefig(os.path.join(FIGS_DIR, "model_samples_moons_conditional_with_.png"))
            plt.close()

flow.load_state_dict(best_state)
# Plot losses as we go
fig, ax1 = plt.subplots(dpi=200)
color = 'crimson'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Training loss', color=color)
ax1.plot(losses_train, label="Training loss", color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'cyan'
ax2.set_ylabel('Validation loss', color=color)  # we already handled the x-label with ax1
ax2.plot(losses_vald, label="Validation loss", color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig(os.path.join(FIGS_DIR, "loss_nvp_moons_with_vald.png"))
plt.show()
```

#### Inference

```python
# Plot the likelihood over data space
X, Y = make_moons(n_data, noise=0.05)
# X = X[Y == curr_class]
X = 2. * (X - X.min()) / (X.max() - X.min()) - 1.

# Get a grid of points spanning the data space
X1, X2 = X.T
dx = 0.1 * (X1.max() - X1.min()) # Same span as data plus a little bit extra
dy = 0.1 * (X2.max() - X2.min())
X_, Y_ = np.mgrid[X1.min() - dx:X1.max() + dx:100j, X2.min() - dy:X2.max() + dy:100j]
XX = torch.as_tensor(np.vstack([X_.ravel(), Y_.ravel()])).float().T

fig, axs = plt.subplots(1, 3, dpi=200, figsize=(7, 3))
tot = np.zeros_like(X_.ravel())
for ax, curr_class in zip(axs[:-1], [-1, 1]):
    X = torch.as_tensor(np.vstack([X_.ravel(), Y_.ravel(), curr_class * np.ones_like(Y_.ravel())])).float().T
    # Flow likelihood
    L_x = np.exp(flow.log_prob(X.to(device)).cpu().detach().numpy())
    ax.scatter(*XX.numpy().T, c=L_x, s=8.)
    ax.set_title(f"Probability of class {(curr_class + 1) // 2}")
    tot += L_x

axs[-1].scatter(*XX.numpy().T, c=tot, s=8.)
axs[-1].set_title("Probability of any class")
plt.savefig(os.path.join(FIGS_DIR, f"likelihood_moons_conditional.png"))
plt.tight_layout()
plt.show()
```
